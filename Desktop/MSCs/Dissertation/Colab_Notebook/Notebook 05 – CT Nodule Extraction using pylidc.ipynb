```python
import pylidc as pl

scans = pl.query(pl.Scan).all()

print("Total scans:", len(scans))
```

##6. Inspect the First CT Scan

```python
scan = scans[0]

print(scan)
```

##7. Display Scan Information

```python
print("Patient ID      :", scan.patient_id)
print("Study UID       :", scan.study_instance_uid)
print("Series UID      :", scan.series_instance_uid)
print("Slice Thickness :", scan.slice_thickness)
print("Pixel Spacing   :", scan.pixel_spacing)
```

##8. Count Nodules

```python
annotations = scan.annotations

print("Number of annotations:", len(annotations))
```

##9. Cluster Radiologist Annotations

```python
clusters = scan.cluster_annotations()

print("Number of nodule clusters:", len(clusters))
```

##10. Show Cluster Sizes

```python
for i, cluster in enumerate(clusters):

    print(
        f"Cluster {i+1}:",
        len(cluster),
        "annotations"
    )
```

##11. Inspect the First Nodule

```python
cluster = clusters[0]

print(cluster)
```

##12. Inspect the First Radiologist Annotation

```python
ann = cluster[0]

print(ann)
```

##13. Display Annotation Characteristics

```python
print("Subtlety      :", ann.subtlety)
print("InternalStructure :", ann.internalStructure)
print("Calcification :", ann.calcification)
print("Sphericity    :", ann.sphericity)
print("Margin        :", ann.margin)
print("Lobulation    :", ann.lobulation)
print("Spiculation   :", ann.spiculation)
print("Texture       :", ann.texture)
print("Malignancy    :", ann.malignancy)
```

##14. Display Bounding Box

```python
bbox = ann.bbox()

print(bbox)
```

##15. Display Centroid

```python
print(ann.centroid)
```
